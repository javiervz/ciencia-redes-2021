# **1. REDES ALEATORIAS**

## **1.1. Define funciones para red aleatoria**

```python
import random
import itertools
import networkx as nx

def _crea_nodos(cantidad_nodos):
    """Entrega una lista de números consecutivos, partiendo 
    en 1, que después serán utilizados como identificadores de 
    los nodos de la red."""
    nodos = list(range(1,cantidad_nodos + 1))
    return nodos

def _iteración_probabilidades(probabilidad_mínima,probabilidad_máxima,cantidad_pasos,verbose=False):
    """Entre una probabilidad mínima y máxima, el ciclo va
    entregando sucesivas probabilidades de conección entre nodos."""
    largo_paso = (probabilidad_máxima - probabilidad_mínima) / cantidad_pasos
    probabilidades = [probabilidad_mínima + (largo_paso * i) for i in range(cantidad_pasos + 1)]
    for probabilidad in probabilidades[:]:
        if verbose: print(f"\nProbabilidad: {round(probabilidad,3)}")
        yield probabilidad

def _crea_red_aleatoria(nodos,probabilidad):
    """A partir de la lista de nodos y una determinada probabilidad
    de conección entre ellos, entrega una instancia de grafo según NetworkX."""
    red = nx.Graph()
    red.add_nodes_from(nodos)
    for nodo_1,nodo_2 in itertools.combinations(nodos,2):
        chance = random.uniform(0,1.0)
        if chance <= probabilidad:
            red.add_edge(nodo_1,nodo_2)
    return red

def _extrae_métricas_red(red,verbose=False):
    """Calcula las métricas de conexión, número de componentes, tamaño, 
    densidad y clustering promedio del componente más grande o de la red."""
    red_conectada = nx.is_connected(red)
    número_componentes_conexas = nx.number_connected_components(red)
    componentes = [red.subgraph(nodos_componente) for nodos_componente in nx.connected_components(red)]
    componentes_vs_tamaño = {componente:componente.size() for componente in componentes}
    componentes_ordenado_tamaño = [componente for componente,tamaño in sorted(componentes_vs_tamaño.items(),
                                                                       key=lambda item: item[1],
                                                                       reverse=True)]
    componente_más_grande = componentes_ordenado_tamaño[0]
    tamaño_componente_más_grande = componente_más_grande.number_of_nodes()
    camino_mínimo_promedio = nx.average_shortest_path_length(componente_más_grande)
    densidad_componente = nx.density(componente_más_grande)
    densidad_red = nx.density(red)
    clustering_promedio_componente = nx.average_clustering(componente_más_grande)
    clustering_promedio_red = nx.average_clustering(red)
    métricas_red = {'Red conectada': red_conectada,
                    'Número componentes': número_componentes_conexas,
                    'Tamaño componente grande': tamaño_componente_más_grande,
                    'Camino mínimo promedio': camino_mínimo_promedio,
                    'Densidad componente': densidad_componente,
                    'Densidad red': densidad_red,
                    'Clustering promedio componente': clustering_promedio_componente,
                    'Clustering promedio red': clustering_promedio_red,
                    'Red': red}
    if verbose:
        for nombre_métrica,métrica_dato in métricas_red.items():
            print(f"\t{nombre_métrica} = {métrica_dato}")
    return métricas_red
```

## **1.2. Simula redes aleatorias**

```python
def _simula_redes_aleatorias(cantidad_nodos,probabilidad_mínima,probabilidad_máxima,cantidad_pasos,verbose=False):
    """Función principal (main) que invoca las funciones anteriores para
    crear las distintas redes aleatorias según los parámetros dados."""
    nodos = _crea_nodos(cantidad_nodos)
    datos_redes_aleatorias = []
    for probabilidad in _iteración_probabilidades(probabilidad_mínima = probabilidad_mínima,
                                                  probabilidad_máxima = probabilidad_máxima,
                                                  cantidad_pasos = cantidad_pasos,
                                                  verbose = verbose):
        red_aleatoria = _crea_red_aleatoria(nodos,probabilidad)
        métricas_red = _extrae_métricas_red(red_aleatoria,verbose=verbose)
        datos_red = {'Probabilidad':probabilidad}
        datos_red.update(métricas_red)
        datos_redes_aleatorias.append(datos_red)
    return datos_redes_aleatorias

cantidad_nodos = 100
probabilidad_mínima = 0.0
probabilidad_máxima = 1.0
cantidad_pasos = 100
verbose = True

datos_redes_aleatorias = _simula_redes_aleatorias(cantidad_nodos = cantidad_nodos,
                                                  probabilidad_mínima = probabilidad_mínima,
                                                  probabilidad_máxima = probabilidad_máxima,
                                                  cantidad_pasos = cantidad_pasos,
                                                  verbose = verbose)
```

## **1.3. Crea Dataframe de resultados**

```python
import pandas as pd

def _crea_dataframe_redes(datos_redes_aleatorias):
    """Con las redes aleatorias creadas anteriormente, entrega
    una instancia de dataframe panda con los datos estructurados."""
    columnas = ['Probabilidad','Red conectada','Número componentes',
                'Tamaño componente grande','Camino mínimo promedio',
                'Densidad componente','Densidad red',
                'Clustering promedio componente','Clustering promedio red',
                'Red']
    df_redes = pd.DataFrame(columns=columnas)
    for datos_red in datos_redes_aleatorias:
        df_redes = df_redes.append(datos_red,ignore_index = True)
    return df_redes

pd.options.display.width = None
pd.options.display.max_rows = None

df_redes_aleatorias = _crea_dataframe_redes(datos_redes_aleatorias)
df_redes_aleatorias
```

## **1.4. Visualiza redes**

```python
import matplotlib.pyplot as plt

def _selecciona_red(df_redes,criterio):
  """Entrega la instancia de grafo aleatorio y su probabilidad
  de conección según un criterio de selección dado."""
  if criterio['Operador'] == '>':
    row = df_redes[df_redes[criterio['Columna']] > criterio['Valor'][0]]
  elif criterio['Operador'] == '<':
    row = df_redes[df_redes[criterio['Columna']] < criterio['Valor'][0]]
  elif criterio['Operador'] == '=':
    row = df_redes[df_redes[criterio['Columna']] == criterio['Valor'][0]]
  if row.shape[0] == 0: print(row)
  row = row.iloc[criterio['Valor'][1]]
  print(row)
  probabilidad = row['Probabilidad']
  red = row['Red']
  return red,probabilidad

def _visualiza_red(red,criterio,probabilidad):
  """Crea un gráfico de la red seleccionada, imprimiendo
  los datos de sus métricas."""
  nx.draw(red, node_size=50)
  título = criterio["Nombre"].upper()
  título += f" (p={round(probabilidad,2)})"
  plt.title(título)
  plt.show()

selección = [{'Nombre': "Varios Componentes", 'Columna': "Número componentes",'Operador':">",'Valor': (20,-1)},
             {'Nombre': "Dos Componentes", 'Columna': "Número componentes",'Operador':">",'Valor': (1,-1)},
             {'Nombre': "Primera Red Conectada", 'Columna': "Red conectada",'Operador':">",'Valor': (0.0,0)},
             {'Nombre': "Clique", 'Columna': "Densidad red",'Operador':"=",'Valor': (1,-1)}]

for criterio in selección:
  red,probabilidad = _selecciona_red(df_redes_aleatorias,criterio)
  _visualiza_red(red,criterio,probabilidad)
```

## **1.5. Visualiza resultados**

```python
def _visualiza_resultados(df_redes,líneas_y,escala_logarítmica):
  """Crea un gráfico –en escala lineal o logarítmica– de una o
  más líneas a partir de los datos de las métricas de todas las
  redes aleatorias guardadas en el dataframe."""
  plt.figure(figsize=(10,3))
  eje_x = df_redes['Probabilidad']
  for línea_y in líneas_y:
    eje_y = df_redes[línea_y]
    plt.plot(eje_x,eje_y)
  plt.xlabel('Probabilidad')
  título = " vs ".join(líneas_y)
  if escala_logarítmica:
    plt.yscale('log')
    título += "  (Logarítimica)"  
  else:
    título += "  (Lineal)"
  plt.title(título)
  plt.legend(líneas_y)
  plt.show()

visualizaciones = [{'Líneas Y': ['Red conectada'], 'Escala logarítmica': False},
                   {'Líneas Y': ['Tamaño componente grande'], 'Escala logarítmica': False},
                   {'Líneas Y': ['Camino mínimo promedio'], 'Escala logarítmica': True},
                   {'Líneas Y': ['Densidad componente','Densidad red'], 'Escala logarítmica': True},
                   {'Líneas Y': ['Clustering promedio componente','Clustering promedio red'], 'Escala logarítmica': True},
                   {'Líneas Y': ['Densidad red','Clustering promedio red'], 'Escala logarítmica': True}]

for visualización in visualizaciones:
  _visualiza_resultados(df_redes = df_redes_aleatorias,
                        líneas_y = visualización['Líneas Y'],
                        escala_logarítmica = visualización['Escala logarítmica'])
```

## **1.6. Conclusiones redes aleatorias**

1. Basta con una probabilidad de conección baja (ca. 0.03-0.04) para que la red aleatoria esté enteramente conectada y, por lo tanto, dejen de existir componentes separados.
2. En una red aleatoria, el camino mínimo promedio decrese rápidamente –*i.e.*, de 6 a 2 pasos– hasta una probabilidad de ca.0.13-0.14. De ahí en adelante (ca.0.15 a 1.0), el camino promedio sigue decresiendo pero marginalmente de 2 a 1 paso.
3. Los valores de densidad y clustering de una red aleatoria difieren en las probabilidades de conección baja (ca.0.0-0.25). De ahí en adelante –*i.e.*, una probabilidad de ca. 0.26 o superior–, la diferencia entre los valores de densidad y clustering de la red aleatoria es marginal (menos de 0.001). Esto probablemente se debe a la ley de los grandes números sumado a que el generador de valores azarosos tiene una distribución uniforme.

# **2. RED LENGUAJES**

## **2.1. Exploración preliminar de red lenguajes**

```python
import pickle

def _exploración_preliminar(red,atributo_peso):
  """Imprime en pantalla los nodos, arcos (junto a sus datos)
  y visualiza el grafo."""
  nodos = red.nodes(data=True)
  print(nodos)
  arcos = red.edges(data=True)
  print(arcos)
  pesos = list(nx.get_edge_attributes(red,name=atributo_peso).values())
  nx.draw(red, 
          node_size=50,
          width=pesos)

nombre_archivo = 'knn-graph-south-america.p'
with open (nombre_archivo,'rb') as file:
  red_lenguajes = pickle.load(file)

_exploración_preliminar(red_lenguajes,'alpha')
```

## **2.2. Crea redes aleatorias lenguajes**

```python
cantidad_lenguajes = red_lenguajes.number_of_nodes()

cantidad_nodos = cantidad_lenguajes
probabilidad_mínima = 0.0
probabilidad_máxima = 1.0
cantidad_pasos = 100
verbose = False

datos_aleatorias_lenguajes = _simula_redes_aleatorias(cantidad_nodos = cantidad_nodos,
                                                      probabilidad_mínima = probabilidad_mínima,
                                                      probabilidad_máxima = probabilidad_máxima,
                                                      cantidad_pasos = cantidad_pasos,
                                                      verbose = verbose)
```

## **2.3. Crea DataFrame Aleatorias Lenguajes**

```python
df_aleatorias_lenguajes = _crea_dataframe_redes(datos_aleatorias_lenguajes)
#df_aleatorias_lenguajes
```

## **2.4. Compara Métricas Real vs. Aleatorias**

```python
def _busca_aleatorio_similar(df_aleatorias,métricas_real,métrica_comparación):
  """Ordena el dataframe de redes aleatorias según su diferencia con una
  determinada métrica real. Entrega la red aleatoria más similar imprimiendo 
  sus métricas."""
  dato_real = métricas_real[métrica_comparación]
  df_ordenado = df_aleatorias.iloc[(df_aleatorias[métrica_comparación]-dato_real).abs().argsort()]
  row_más_similar = df_ordenado[:1]
  probabilidad = row_más_similar['Probabilidad'].values[0]
  métrica_aleatoria = row_más_similar[métrica_comparación].values[0]
  print(f"\nComparación: {métrica_comparación}")
  print(f"\tProbabilidad: {probabilidad}")
  print(f"\tMétrica Real: {dato_real}")
  print(f"\tMétrica Aleatoria: {métrica_aleatoria}")
  red_similar = row_más_similar['Red'].values[0]
  return red_similar

print(f"Métricas Reales Lenguajes:")
métricas_lenguajes = _extrae_métricas_red(red_lenguajes,verbose=True)

comparaciones = ['Camino mínimo promedio',
                 'Densidad red',
                 'Clustering promedio red']

for comparación in comparaciones:
  red_aleatoria_similar = _busca_aleatorio_similar(df_aleatorias_lenguajes,
                                                   métricas_lenguajes,
                                                   comparación)
```

## **2.5. Conclusiones red lenguaje**

1. Comparando el **camino mínimo promedio**, la red de lenguaje se asemeja más a un red aleatoria con una probabilidad de conecciones de **p = 0.03**. En esta métrica, la red de lenguaje tiene un valor de **5.018**, mientras que la red aleatoria de **4.430**.

2. Comparando la **densidad de la red**, la red de lenguaje se asemeja más a un red aleatoria con una probabilidad de conecciones de **p = 0.10**. En esta métrica, la red de lenguaje tiene un valor de **0.095**, mientras que la red aleatoria de **0.098**.

3. Comparando el **clustering promedio de la red**, la red de lenguaje se asemeja más a un red aleatoria con una probabilidad de conecciones de **p = 0.72**. En esta métrica, la red de lenguaje tiene un valor de **0.727**, mientras que la red aleatoria de **0.721**.

4. En una red aleatoria, como vimos, el clustering promedio y la densidad tienden a ser iguales (*supra* 1.6.3). En esta red de lenguaje, en cambio, clustering y densidad difieren en **0.632** (0.727 – 0.095).

5. Si los valores de las métricas de una red real se asemejaran a los de una y la misma red aleatoria con el mismo valor p, entonces cabría sospechar que las conexiones en la red real son aleatorias también (o uniformemente distribuidas).

6. Esta red de lenguaje –por el contratio – tiene valores de "camino mínimo promedio", "densidad" y "clustering promedio" que corresponden a **distintas redes aleatorias** con a su vez distintas probabilidades de conección: p=0.03,0.10 y 0.72 respectivamente.

7. Tal diferencia con las redes aleatorias permite concluir que la representación en red de las relaciones entre estos lenguajes sí captura información.

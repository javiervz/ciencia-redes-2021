### Calcule el número de componentes conexas en función de  p :

Generamos 100 grafos y calculamos el número de componentes conexas para cada uno, almacenando los datos en una lista

```python
p_list=np.random.rand(100)
p_list=sorted(p_list)
```

```python
n_comp_list=[]
diam_list=[]
av_short_list=[]
den_list=[]
clust_list=[]
```

```python
for p in p_list:
  graph=nx.erdos_renyi_graph(100, p)
  n_comp_list.append(nx.number_connected_components(graph))
  clust_list.append(nx.average_clustering(graph))
  den_list.append(nx.density(graph))
  if nx.is_connected(graph):
    diam_list.append(nx.diameter(graph))
    av_short_list.append(nx.average_shortest_path_length(graph))
  else:
    S = graph.subgraph(max(nx.connected_components(graph), key=len)).copy()
    diam_list.append(nx.diameter(S))
    av_short_list.append(nx.average_shortest_path_length(S))
```

Calcule el número de componentes conexas en función de p :

```python
a=pd.DataFrame({'p':p_list})
b=pd.DataFrame({'n° comp':n_comp_list})
c=pd.DataFrame({'diametro':diam_list})
d=pd.DataFrame({'camino min':av_short_list})
e=pd.DataFrame({'densidad':den_list})
f=pd.DataFrame({'clust':clust_list})
df=pd.concat([a,b,c,d,e,f],axis=1)
```

```python
df
```

```python
plt.plot(p_list,diam_list)
plt.xlabel("p_completos")
plt.ylabel("Diámetro")
```

```python
plt.plot(p_list,clust_list)
plt.xlabel("p")
plt.ylabel("Average Clustering")
```

```python
plt.plot(p_list,den_list)
plt.xlabel("p")
plt.ylabel("Densidad")
```

Conclusiones

- A partir de un p de 7%, obtenemos grafos conectados, es decir, que consisten en una sola componente conexa
-Los grafos generados con un valor de p mayor, tienen un diámetro de 2, un tamañó mínimo promedio de la mayor (única) componente conexa cercano a 1; y presentan valores de Densidad y Coeficiente de clustering muy similares
-El valor de Coeficiente de Clustering y Densidad del grafo aumenta aproximadamente en forma lineal con el aumento del valor de p

2. Un **grafo real!** El archivo "knn-graph-south-america.p" contiene en formato pickle (pueden usar "pickle.load(open('knn-graph-south-america.p','rb'))" para abrirlo) un grafo, que llamaremos $G_{SA}$, construido de la siguiente forma:

- A partir de datos [tipológicos](https://sails.clld.org), obtuvimos información acerca de 88 lenguas de Sudamérica. Cada lengua es representada por 9 rasgos. 
- Con una famosa definición de [distancia](https://en.wikipedia.org/wiki/Hamming_distance), determinamos los $k$-vecinos cercanos de cada lengua. Fijamos $k=5$.
- Con esto, cada lengua se conecta con las $k$ lenguas más cercanas.

Realice los siguientes cálculos:

- Construya 100 grafos $G_{n,p}$, con $n$ igual al número de nodos de $G_{SA}$, y distintos valores de $p$.
- ¿Qué valor de $p$ entrega el grafo aleatorio que más se asemeja a $G_{SA}$? Elija dos medidas de las usadas en la pregunta 1. 
- Concluya de forma simple y precisa.

Definimos una función que nos permitirá comparar el Coeficiente de clustering y la Densidad de los grafos más adelante :

```python
def diferencia(x,y):
  dif=abs(x-y)
  return dif
```

```python
data=pickle.load(open('knn-graph-south-america.p','rb'))
nx.draw(data,node_size=50,node_color="blue")
```

Examinamos las características del grafo :

```python
print("El número de nodos es: ", len(data))
print("El número de componentes conexas es: ", nx.number_connected_components(data))
print("El tamaño de la componente conexa más grande es: ", nx.diameter(data))
print("El camino mínimo promedio de la componente conexa más grande es: ", nx.average_shortest_path_length(data))
print("La densidad es: ", nx.density(data))
print("El coeficiente de clustering promedio es: ", nx.average_clustering(data))
```

```python
data_p_list=np.random.rand(len(data))
data_p_list=sorted(data_p_list)
```

```python
data_n_comp_list=[]
data_diam_list=[]
data_av_short_list=[]
data_den_list=[]
data_clust_list=[]
dif_den={}
dif_clust={}
```

```python
for p in data_p_list:
  graph2=nx.erdos_renyi_graph(len(data), p)
  data_n_comp_list.append(nx.number_connected_components(graph2))
  data_clust_list.append(nx.average_clustering(graph2))
  data_den_list.append(nx.density(graph2))
  if nx.is_connected(graph2):
    data_diam_list.append(nx.diameter(graph2))
    data_av_short_list.append(nx.average_shortest_path_length(graph2))
  else:
    S = graph2.subgraph(max(nx.connected_components(graph2), key=len)).copy()
    data_diam_list.append(nx.diameter(S))
    data_av_short_list.append(nx.average_shortest_path_length(S))

  dif_den[diferencia(nx.density(graph2),nx.density(data))]=p
  dif_clust[diferencia(nx.average_clustering(graph2),nx.average_clustering(data))]=p
```

```python
densidades = list(dif_den.items())
densidades=sorted(densidades)
```

```python
coefs_clust = list(dif_clust.items())
coefs_clust=sorted(coefs_clust)
```

```python
aa=pd.DataFrame({'p':data_p_list})
bb=pd.DataFrame({'n° comp':data_n_comp_list})
cc=pd.DataFrame({'diametro':data_diam_list})
dd=pd.DataFrame({'camino min':data_av_short_list})
ee=pd.DataFrame({'densidad':data_den_list})
ff=pd.DataFrame({'clust':data_clust_list})
dff=pd.concat([aa,bb,cc,dd,ee,ff],axis=1)
```

```python
dff
```

Finalmente presentamos aquel $p$ que entrega el grafo aleatorio que más se asemeja a $G_{SA}$. Esto lo hacemos comparando los Coeficiente de clustering y la Densidad entre el grafo generado en nuestra iteración y el grafo $G_{SA}$

```python
print("La diferencia de los Coeficientes de Clustering entre el grafo generado y el grafo Gsa, junto con el p correspondiente, son, respectivamente:", coefs_clust[0])
```

```python
print("La diferencia de las Densidades entre el grafo generado y el grafo Gsa, junto con el p correspondiente, son, respectivamente:", densidades[0])
```

Conclusiones

- La generación de grafos con 88 nodos (la misma cantidad de nodos del grafo de referencia) describe el mismo comportamiento de los grafos generados con 100 nodos
-Aquel p con el cual se obtiene el grafo que presenta mayor similitud con el grafo de referencia, teniendo en cuenta como parámetro de comparación la Densidad del grafo, es : 0.0915947663082528. Es decir, un grafo en el cual cada uno de sus nodos tiene una probabilidad aproximada de 9% de crear enlaces con cada uno del resto de nodos.
-Aquel p con el cual se obtiene el grafo que presenta mayor similitud con el grafo de referencia, teniendo en cuenta como parámetro de comparación la el Coeficiente de Clustering promedio del grafo, es : 0.7146451616917352. Es decir, un grafo en el cual cada uno de sus nodos tiene una probabilidad aproximada de 7% de crear enlaces con cada uno del resto de nodos.

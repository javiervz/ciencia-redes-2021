```python
#Librerias
import networkx as nx
import random
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import numpy as np
```

## Primera Parte

Modelo $G_{n,p}$. En el modelo **Erdős-Rényi**, cada par de nodos se une a través de una arista con una probabilidad $p$. Considere un conjunto de nodos $V$ de tamaño $n=100$. Considere 100 valores para $p$ entre 0 y 1.

```python
#Constantes
n = 100
```

### Implemente el modelo $G_{n,p}$

```python
#Funcion que devuelve un grafo aleatorio (modelo de Paul Erdős y Alfréd Rényi)
def erdos_renyi_graph(n,p):
    #Creacion del grafo vacio
    G = nx.Graph()
 
    #Se generan los n cantidad de nodos
    for node in range(n):
        G.add_node(node)

    node_list = list(G.nodes)
    
    #Se recorren todas las combinaciones de nodos
    for i, node1 in enumerate(node_list):
        for j, node2 in enumerate(node_list):
            if i<j:
                #Si la chance esta dentro de la probabilidad entonces se genera un enlace
                chance = random.random()
                if chance <= p:
                    G.add_edge(node1,node2)
    return G

#Funcion para obtener el promedio de grado de un grafo
def average_degree(graph):
    average_degree = 0
    for node, degree in graph.degree():
        average_degree = average_degree + degree
        
    return (average_degree/nx.number_of_nodes(graph))
```

### Dibuje grafos para diferentes valores del parámetro $p$ (3-5 ejemplos)

**Probaremos con probabilidad igual 0%**

```python
#Probabilidad 0%
p = 0
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

Al tener 0% de probabilidad de generar enlaces, nos queda un grafico discreto, sin densidad en donde no existen aristas ni relaciones

**Probaremos una probabilidad baja (5%) para visualizar enlaces**

```python
#Probabilidad 5%
p = 0.05
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

Se generaron nodos con 1 o 3 enlaces sabiendo que tienen el potencial de tener 99 enlaces

**Probaremos con probabilidad igual 100%**

```python
#Probabilidad 100%
p = 1
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

Al tener 100% de probabilidad de generar enlaces, nos queda un grafico completo, muy denso en donde todos los nodos se comunican entre si

**Otras probabilidades**

```python
#Probabilidad 50%
p = 0.5
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

```python
#Probabilidad aleatoria
p = random.random()
print("Probabilidad Aleatoria: ", p*100,"%")
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

### Calcule el **número de componentes conexas** en función de $p$.

```python
#Probabilidad de 4% para generar varios componentes
p = 0.04
G = erdos_renyi_graph(n,p)
pos = nx.spring_layout(G)
nx.draw(G,pos)
```

```python
#nx.number_connected_components(G) devolverá la cantidad de componentes conexas del grafo
number_connected_components = nx.number_connected_components(G)

print("Cantidad de componentes conexas: ",number_connected_components)
```

### Calcule el **tamaño de la componente conexa más grande** en función de $p$. 

```python
#nx.connected_components(G) devolverá los componentes conexos
#max_length,longest_component = max([(len(i),i) for i in list(nx.connected_components(G))])
#longest_component = max(nx.connected_components(G), key=len)
#max_length = len(max(nx.connected_components(G), key=len))

#Obtenemos un nuevo Grafo a partir de un subgrafo de G
S = G.subgraph(max(nx.connected_components(G), key=len)).copy()

max_length = nx.number_of_nodes(S)

print("La componente mas grande tiene un tamaño de: ",max_length)
```

### Calcule el **camino mínimo promedio de la componente conexa más grande** en función de $p$.

```python
average_shortest_path = nx.average_shortest_path_length(S)

print("El promedio de caminos mas corto es de: ",average_shortest_path)
```

### Calcule la **densidad** en función de $p$.

```python
densidad = nx.density(G)

print("La densidad del grafo es:",densidad)
```

### Calcule el **coeficiente de clustering promedio** en función de $p$.

```python
average_clustering_coef = nx.average_clustering(G)

print("La probablidad promedio de que 2 vecinos de un mismo nodo sean vecinos entre si (Clustering Coefficient) es:",(average_clustering_coef*100),"%")
```

### Conclusiones

En los ejercicios anteriores se pudo notar como la probabilidad (p) afecta (casi de manera proporcional) los diferentes caracteristicas de los grafos.
Por ejemplo, a menor **p** el grafo tiende a ser discreto generando mas componentes conexas, mientras que a mayor **p** el grafo tiende a ser completo y generar menos componentes


## Segunda Parte

Un **grafo real!** El archivo "knn-graph-south-america.p" contiene en formato pickle (pueden usar "pickle.load(open('knn-graph-south-america.p','rb'))" para abrirlo) un grafo, que llamaremos $G_{SA}$, construido de la siguiente forma:

- A partir de datos [tipológicos](https://sails.clld.org), obtuvimos información acerca de 88 lenguas de Sudamérica. Cada lengua es representada por 9 rasgos. 
- Con una famosa definición de [distancia](https://en.wikipedia.org/wiki/Hamming_distance), determinamos los $k$-vecinos cercanos de cada lengua. Fijamos $k=5$.
- Con esto, cada lengua se conecta con las $k$ lenguas más cercanas.

```python
H = pickle.load(open('knn-graph-south-america.p','rb'))
```

```python
plt.figure(figsize=(18,12))
pos = nx.spring_layout(H)
nx.draw(H,pos)
```

### Construya 100 grafos $G_{n,p}$, con $n$ igual al número de nodos de $G_{SA}$, y distintos valores de $p$.

```python
n = nx.number_of_nodes(H)
```

```python
#Genero 101 grafos aleatorios
graph_list = []

for i in range(101):
    prob = i/100
    graph_dict = {'id': i, 'graph': erdos_renyi_graph(n,prob), 'p': prob}
    graph_list.append(graph_dict)
```

### ¿Qué valor de $p$ entrega el grafo aleatorio que más se asemeja a $G_{SA}$? Elija dos medidas de las usadas en la pregunta 1.

```python
#Obtengo las metricas del Graph de lenguas de sudamerica
sa_graph_num_connected_components = nx.number_connected_components(H)
sa_graph_avg_shortest_path = nx.average_shortest_path_length(H.subgraph(max(nx.connected_components(H), key=len)).copy())
sa_graph_density = nx.density(H)
sa_graph_avg_clustering_coef = nx.average_clustering(H)
sa_graph_avg_degree = average_degree(H)
print('--Caracteristicas Grafo de Sudamerica--')
print('Cantidad de componentes conexas:',sa_graph_num_connected_components)
print('Promedio de caminos mas cortos:',sa_graph_avg_shortest_path)
print('Densidad del grafo:',sa_graph_density)
print('Promedio de coeficiente de clustering:',sa_graph_avg_clustering_coef)
print('Promedio de grado del grafo:',sa_graph_avg_degree)
```

```python
#Obtengo las caracteristicas de los grafos aleatorios generados y las guardo en un dataframe
column_names = ["id", "probability", "num_connected_components", "avg_shortest_path", "density", "avg_clustering_coef"]
graph_comparison_df = pd.DataFrame(columns = column_names)

for i in graph_list:
    num_connected_components = nx.number_connected_components(i['graph'])
    avg_shortest_path = nx.average_shortest_path_length(i['graph'].subgraph(max(nx.connected_components(i['graph']), key=len)).copy())
    density = nx.density(i['graph'])
    avg_clustering_coef = nx.average_clustering(i['graph'])
    avg_degree = average_degree(i['graph'])
    
    graph_comparison_df = graph_comparison_df.append({'id': int(i['id']), 'probability':i['p'], 'num_connected_components':num_connected_components,'avg_shortest_path':avg_shortest_path, 'density':density, 'avg_clustering_coef':avg_clustering_coef, 'avg_degree':avg_degree}, ignore_index=True)
    
graph_comparison_df
```

```python
#Elimino las filas que tengan diferente numero de componentes que el grafico de sur america
graph_comparison_df = graph_comparison_df.drop(graph_comparison_df[graph_comparison_df.num_connected_components != sa_graph_num_connected_components].index)
graph_comparison_df
```

```python
#Calculo la diferencia de cada caracteristica (prefix:diff_) entre cada grafo aleatorio y el grafo de sur america
#graph_comparison_df['diff_num_connected_components'] = sa_graph_num_connected_components - graph_comparison_df['num_connected_components']
graph_comparison_df['diff_avg_shortest_path'] = abs(sa_graph_avg_shortest_path - graph_comparison_df['avg_shortest_path'])
graph_comparison_df['diff_density'] = abs(sa_graph_density - graph_comparison_df['density'])
graph_comparison_df['diff_avg_clustering_coef'] = abs(sa_graph_avg_clustering_coef - graph_comparison_df['avg_clustering_coef'])
graph_comparison_df['diff_avg_degree'] = abs(sa_graph_avg_degree - graph_comparison_df['avg_degree'])
graph_comparison_df
```

```python
#Promedio de las diferencias
graph_comparison_df['diferencia'] = abs((graph_comparison_df['diff_avg_shortest_path'] + graph_comparison_df['diff_density'] + graph_comparison_df['diff_avg_clustering_coef'] + graph_comparison_df['diff_avg_degree'])/4)
```

```python
#Primer Experimento. Nos basamos en el promedio de todas las diferencias de todas las caracteristicas (A excepcion de numero de componentes conexas que fueron filtradas en pasos anteriores).
#Orden de la tabla ascendente a partir de la diferencia
graph_comparison_df.sort_values(by='diferencia', ascending=True)
```

A medida que el resultado de la diferencia se aproxima a 0, significaria que son mas parecidos

```python
#Grafo de probabilidad 9%
plt.figure(figsize=(18,12))
U = graph_list[9]['graph']
pos = nx.spring_layout(U)
nx.draw(U,pos)
```

```python
#Segundo Experimento, basandonos en 2 caracteristicas (Numero de componentes = 1 y diferencia en el promedio de coeficiente de clustering)
#Orden de la tabla ascendente a partir de la caracteristica del promedio del coeficiente de clustering
graph_comparison_df.sort_values(by='diff_avg_clustering_coef', ascending=True)
```

Luego se obtuvo el valor absoluto de la resta de cada metricas con las del grafo de sudamerica correspondiente para cuantificar cada diferencia.

## Implementando el modelo $G_{N,p}$

Para este modelo se debe generar un grafo con $N$ nodos, con una probabilidad $p$ de que cada par de nodos esten conectados.

La cantidad de pares de nodos es $\frac{N(N-1)}{2}$

---
Para este ejercicio, llenarémos el grafo `g` con 100 nodos, identificados con números entre 0 y 99.


Teniendo claro cómo llenar el grafo, definirémos la función que generará el modelo de **Erdős-Rényi**

```python
# N -> cantidad de nodos
# p -> probabilidad [0,1]

def G(N,p):
    # genero el grafo
    graph = nx.Graph()
    
    # AGREGO EL PRIMER NODO
    graph.add_node(0)
    
    for i in range(N-1):
        graph.add_node(i+1)          #AGREGO UN NUEVO NODO
        for j in range(i+1,N-1):
            if p >= random.random():
                graph.add_edge(i,j)  #CONECTO EL NODO i CON SU PAR j 
    return graph
```

De esta forma, se recorre media matriz sin pasar por la diagonal. Ejemplo para N=5

\begin{bmatrix}
   & * & * & * & * \\
   &   & * & * & * \\
   &   &   & * & * \\
   &   &   &   & * \\
   &   &   &   &   \\
 \end{bmatrix}
 
 Recorriendo $N(N-1)/2$ casillas, o pares de nodos
 
 ---
 
Para poner a **prueba el modelo creado**, se comprobará lo siguiente:
- que el número real de conexiones se acerque al estimado $\langle L \rangle = p \cdot \frac{N(N-1)}{2}$
- que el grado promedio obtenido se acerque a $\langle k \rangle = \frac{2 \langle L \rangle}{N} = p(N-1)$




```python
n = 1000
p = 1/n

g = G(n,p)

pares = n*(n-1)/2

L = p*pares

k = 2*L/n
```

```python
print(g.number_of_edges(), L)
```

```python
suma = 0
for i,d in g.degree():
    suma += d
    
print(suma/n, k)
```

Luego de varios experimentos, se observa que ambas premisas a comprobar se cumplen.

## Análisis de grafos

Para cada uno de esos conjuntos se debe realizar una serie de análisis, según el valor de $p$:
- Calcule el **número de componentes conexas** en función de $p$.
- Calcule el **tamaño de la componente conexa más grande** en función de $p$.
- Calcule el **camino mínimo promedio de la componente conexa más grande** en función de $p$.
- Calcule la **densidad** en función de $p$.
- Calcule el **coeficiente de clustering promedio** en función de $p$.


### 1. Número de CC

En primer lugar, se debe obtener el número de componentes conexas

```python
def cantidad_cc(graph):
    cc = list(nx.connected_components(graph))
    return len(cc)
# Devuelve un entero con la cantidad de cc
```

### 2. Tamaño de la CC más grande

Luego calculamos el tamaño de la componente conexa (CC) más grande presente en el grafo

```python
def largest_cc_size(graph):
    largest_cc = max(nx.connected_components(graph), key=len)
    return len(largest_cc), largest_cc
```

La función `largest_cc_size` retorna una tupla con:
- cantidad de nodos de la CC más grande
- lista con los nodos pertenecientes a dicha CC

---

### 3. Camino mínimo promedio

Teniendo la CC más grande, buscaremos entre cada par de nodos, el camino más corto que existe entre ellos y calcularemos su promedio.

```python
def average_shortest_path(graph):
    ls = list(graph.nodes())
    l = []
    bool_cc = True
    for i in range(len(ls)):
        for j in range(i+1,len(ls)):
            try:
                l.append(nx.shortest_path_length(graph,source=ls[i],target=ls[j]))            
            except nx.NetworkXNoPath:
                continue
    if 0 in l:
        bool_cc = False    # Si hay un cero en la lista, el grafo presenta más de una CC
    
    return sum(l)/len(l)
```

La función `average_shortest_path` retorna el tamaño promedio de los caminos más cortos entre cada par de nodos. La libreria nx ofrece una función, pero devuelve error cuando el grafo ingresado no tiene conexiones. En la función propia se concidera esa excepción.

---

### 4. Densidad del grafo

Luego se pide obtener la densidad del grafo, en función de $p$. Para esto utilizaremos la función `nx.density()` que facilida la libreria nx.

---

### 5. Coeficiente de clustering promedio

También se solicita el coeficiente de clustering promedio de cada grafo. Para esto se utilizará la función `nx.average_clustering_coeficent()`. 

---

### Experimentos con distintos conjuntos de $p$

Para las siguientes actividades, se necesita una cierta cantidad de grafos, diferenciadas por el valor de $p$. Lo interesante es comparar las distintas distribuciones que podría tener el conjunto de valores de $p$.

Algunas distribuciones podrian ser:
- $A = \{ p_n \rightarrow rand(0,1) \mid 1\le n \le 100 , n \in \mathbb{N}\}$
- $B = \{ \frac{n}{100} \mid 0 \le n < 100 , n \in \mathbb{N}\}$


```python
def analisis_grafo(C_p, title):
    cant_cc = []
    max_cc = []
    min_path = []
    densidad = []
    coef_clust = []
    N = 100
    for p in C_p:
        b_cc_size = 1            #tamaño de la cc mas grande

        # genero el grafo
        graph = G(N, p)

        #cantidad de componentes conexas
        cant_cc.append(cantidad_cc(graph))

        #busco la componente más grande
        b_cc_size, largest_cc = largest_cc_size(graph)
        max_cc.append(b_cc_size)
        
        lcc = graph.subgraph(largest_cc).copy()

        # Busco el camino mínimo promedio de la cc mas grande
        min_path.append(average_shortest_path(lcc))

        #densidad
        densidad.append(nx.density(graph))

        #coeficiente clusterin promedio
        coef_clust.append(nx.average_clustering(graph))
        
    plt.plot(C_p,cant_cc)
    plt.title('Cantidad de componentes conexas - ' + title)
    plt.xlabel('p')
    plt.show()

    plt.plot(C_p,max_cc)
    plt.title('Tamaño de CC más agrande - ' + title)
    plt.xlabel('p')
    plt.show()

    plt.plot(C_p,min_path)
    plt.title('Tamaño del camino más corto en la mayor CC - ' + title)
    plt.xlabel('p')
    plt.show()

    plt.plot(C_p,densidad)
    plt.title('Densidad del grafo - ' + title)
    plt.xlabel('p')
    plt.show()

    plt.plot(C_p,coef_clust)
    plt.title('Coeficiente de clustering promedio - ' + title)
    plt.xlabel('p')
    plt.show()
```


---

#### Conjunto $A$

```python
A = np.random.random(size=100)
A.sort()

analisis_grafo(A,'Conjunto A')
```

---

#### Conjunto $B$

```python
B = np.arange(0.01,1.01,0.01)

analisis_grafo(B,'Conjunto B')
```

### Conclusión

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
import pickle
```

```python
G_sa = pickle.load(open('knn-graph-south-america.p','rb'))
```

Analisarémos algunas características de $G_{SA}$

1. El **coeficiente de cluster** promedio

```python
G_sa_coef_clust = nx.average_clustering(G_sa)
print(G_sa_coef_clust)
```

2. **Tamaño minimo promedio** de la CC más grande de $G_{SA}$

```python
G_sa_max_cc , largest_cc = largest_cc_size(G_sa)

lcc_gsa = G_sa.subgraph(largest_cc).copy()

G_sa_min_path = average_shortest_path(lcc_gsa)

print(G_sa_min_path)
```

A continuación, se compararán los valores de las características anteriores con 100 grafos aleatorios, generados con distintos valores de $p$.

```python
A = np.random.random(size=100)
A.sort()
```

```python
N = G_sa.number_of_nodes()

dif_min_path   = []
dif_coef_clust = []
for p in A:
    
    b_cc_size    = 0            #tamaño de la cc mas grande
    g_min_path   = 0
    g_coef_clust = 0

    # genero el grafo
    graph = G(N, p)
    
    b_cc_size, largest_cc = largest_cc_size(graph)

    lcc = graph.subgraph(largest_cc).copy()

    # Busco el camino mínimo promedio de la cc mas grande
    g_min_path = average_shortest_path(lcc)

    #coeficiente clusterin promedio
    g_coef_clust = nx.average_clustering(graph)
    
    dif_min_path.append(abs(g_min_path - G_sa_min_path))
    
    dif_coef_clust.append(abs(g_coef_clust - G_sa_coef_clust))
```

Con esto podremos obtener el valor de $p$ que genera un grafo más parecido a $G_{SA}$, según ambas características obtenidas anteriormente.

---

Para el **tamaño mínimo promedio**, se obtiene. lo siguiente:

```python
index_min_path = min(range(len(dif_min_path)), key=dif_min_path.__getitem__)

print("Iteración: " + str(index_min_path))
print("Valor de p: " + str(A[index_min_path]))
print("Diferencia con G_sa: " + str(dif_min_path[index_min_path]))
```

```python
plt.plot(A,dif_min_path)
plt.title("Diferencia en el tamaño mínimo promedio")
plt.xlabel("p")
plt.show()
```

Para el **coeficiente de clustering** se obtiene lo siguiente

```python
index_coef_clust = min(range(len(dif_coef_clust)), key=dif_coef_clust.__getitem__)

print("Iteración: " + str(index_coef_clust))
print("Valor de p: " + str(A[index_coef_clust]))
print("Diferencia con G_sa: " + str(dif_coef_clust[index_coef_clust]))
```

```python
plt.plot(A,dif_coef_clust)
plt.title("Diferencia en el coeficiente de clustering")
plt.xlabel("p")
plt.show()
```

```python
import camelot
```

```python
tables1 = camelot.read_pdf('datos/1.pdf')
tables2 = camelot.read_pdf('datos/2.pdf')
tables3 = camelot.read_pdf('datos/3.pdf')
tables4 = camelot.read_pdf('datos/4.pdf')
tables5 = camelot.read_pdf('datos/5.pdf')
tables6 = camelot.read_pdf('datos/6.pdf')
tables7 = camelot.read_pdf('datos/7.pdf')
tables8 = camelot.read_pdf('datos/8.pdf')
tables9 = camelot.read_pdf('datos/9.pdf')
tables10 = camelot.read_pdf('datos/10.pdf')
```

```python
for T in [tables1,tables2,tables3,tables4,tables5,tables6,tables7,tables8,tables9,tables10]:
    print("Total tables extracted:", T.n)
```

```python
frames = []
for T in [tables1,tables2,tables3,tables4,tables5,tables6,tables7,tables8,tables9,tables10]:
    frames+=[T[0].df]
```

```python
import pandas as pd
```

```python
result = pd.concat(frames)
```

```python
result.columns = result.iloc[0]
result = result[1:]
```

```python
result
```

```python
comunidades = set(result['Community'])
```

```python
comunidades
```

```python
dict_com_species_services = {}

for com in comunidades:
    result_com = result[result['Community']==com]
    dict_species_services = dict(zip(result_com['Plant_species'],result_com['Plant_service']))
    dict_com_species_services[com]=dict_species_services
```

```python
dict_com_species_services['Zabalo']
```

### construyamos redes bipartitas!!!

```python
import networkx as nx
from networkx.algorithms import bipartite

redes_bipartitas = {}

for com in comunidades:
    
    B = nx.Graph()
    B.add_nodes_from(dict_com_species_services[com].keys(), bipartite=0)
    B.add_nodes_from(dict_com_species_services[com].values(), bipartite=1)
    B.add_edges_from([(key,dict_com_species_services[com][key]) for key in dict_com_species_services[com].keys()])
    redes_bipartitas[com]=B
```

```python
redes_bipartitas
```

```python
import matplotlib.pyplot as plt

G=redes_bipartitas['Villanueva']

species = dict_com_species_services[com].keys()
services = dict_com_species_services[com].values()
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
s = G.subgraph(Gcc[0])

fig, ax = plt.subplots(dpi=800)
labels={i:i for i in s.nodes}
pos = nx.kamada_kawai_layout(s)
species = [node for node in species if node in s.nodes()]
services = [node for node in services if node in s.nodes()]
nx.draw_networkx_nodes(s, pos, nodelist=species,node_size = 25, node_color='gold') 
nx.draw_networkx_nodes(s, pos, nodelist=services,node_size = 25, node_color='m')
nx.draw_networkx_edges(s, pos, alpha=0.5,width=0.5,edge_color='k')
nx.draw_networkx_labels(s,pos,labels,alpha=0.75,font_size=5,font_color='k')
plt.axis('off')
plt.show()
```
